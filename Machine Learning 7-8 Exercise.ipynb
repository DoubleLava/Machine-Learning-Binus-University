You are a data scientist tasked with analyzing the Titanic dataset.

Your goal is to build a logistic regression model that predicts whether a passenger survived the Titanic disaster.

Dataset link:ðŸ”— https://www.kaggle.com/competitions/titanic

## Step 1. Data Preparation

Goal: Load data, explore it, select features, handle missing values, encode categorical data.

```python
# Step 1: Data preparation

import pandas as pd
import numpy as np
```

```python
# Load the Titanic dataset
data = pd.read_csv("train.csv")

# Preview the dataset
print(data.head())
print(data.info())
```

```python
# Target variable (0 = did not survive, 1 = survived)
y = data["Survived"]

# Select relevant features
# We choose simple and commonly used features
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]

X = data[features]
print(X.head())
```

```python
# Check missing values
print(X.isnull().sum())

# Fill missing Age values with median
X.loc[:, "Age"] = X["Age"].fillna(X["Age"].median())

# Convert categorical variable "Sex" into numerical
X.loc[:, "Sex"] = X["Sex"].map({"male": 0, "female": 1})
```

## Step 2. Trainâ€“Test Split

Goal: Split the dataset into 70% training and 30% testing.

```python
# Step 2: Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y  # keeps class balance
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

## Step 3. Model Development

Goal: Build and train a logistic regression model.

```python
# Step 3: Model development

from sklearn.linear_model import LogisticRegression

# Initialize logistic regression model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)

# Display model coefficients
coefficients = pd.Series(model.coef_[0], index=features)
print("Model coefficients:")
print(coefficients)
```

### Logistic Regression Equation

#### (1) Linear combination (logit)

For a passenger with features:

ð‘¥ = (Pclass, Sex, Age, SibSp, Parch, Fare)

the model first computes:

ð‘§ = ð›½0 + ð›½1(Pclass) + ð›½2(Sex) + ð›½3(Age) + ð›½4(SibSp) + ð›½5(Parch) + ð›½6(Fare)

#### Logistic (sigmoid) function

The linear output is converted into a probability of survival:

P(Survived=1|x) = ðœŽ(z) = 1 / (1 + e^-z)

#### Decision rule (classification)

y=1 if P(Survived=1|x) >= 0.5

y=0 if P(Survived=1|x) < 0.5

## Step 4. Model Evaluation

Goal: Evaluate model using Accuracy, Precision, Recall, ROC-AUC.

```python
# Step 4: Model evaluation

from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score

# Predict class labels
y_pred = model.predict(X_test)

# Predict probabilities for ROC-AUC
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("Model performance:")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("ROC-AUC  :", roc_auc)
```

## Step 5. Interpretation and Insights

Goal: Analyze the confusion matrix and interpret TP, FP, TN, FN.

```python
# Step 5: Interpretation and insights

from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Extract values
TN, FP, FN, TP = cm.ravel()

print("Confusion Matrix:")
print(cm)

print("\nInterpretation:")
print("True Positives (TP):", TP, "-> Correctly predicted survivors")
print("False Positives (FP):", FP, "-> Predicted survived but did not")
print("True Negatives (TN):", TN, "-> Correctly predicted non-survivors")
print("False Negatives (FN):", FN, "-> Predicted non-survivor but survived")
```

### Insights:
- High TP means the model is good at identifying survivors
- High FN means missed survivors (important in safety contexts)
- Logistic regression provides interpretable coefficients
- Model assumes linear relationship in log-odds
Limitations:
- Cannot capture complex nonlinear patterns
- Performance depends heavily on feature selection
